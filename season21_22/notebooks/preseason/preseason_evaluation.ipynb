## Predict points for given time horizon of future gameweeks

```python
fpl_fixtures_data = json.loads(requests.get('https://fantasy.premierleague.com/api/fixtures/').text)
fpl_fixtures = pandas.DataFrame(fpl_fixtures_data)
fpl_fixtures
```

```python
fpl_fixtures['home team'] = [teams[i] for i in fpl_fixtures['team_h']-1]
fpl_fixtures['away team'] = [teams[i] for i in fpl_fixtures['team_a']-1]
fpl_fixtures[fpl_fixtures['event']==1]
```

```python
path = Path('../../data/fbref/team_stats_for20_21.csv')
team_stats_for = pandas.read_csv(path, index_col=0, header=[0,1])
team_stats_for.rename(columns={'Unnamed: 0_level_0':'General','Unnamed: 1_level_0':'General',\
                               'Unnamed: 2_level_0':'General','Unnamed: 3_level_0':'General'}, inplace=True)
team_stats_for
```

```python
path = Path('../../data/fbref/team_stats_against20_21.csv')
team_stats_against = pandas.read_csv(path, index_col=0, header=[0,1])
team_stats_against.rename(columns={'Unnamed: 0_level_0':'General','Unnamed: 1_level_0':'General',\
                               'Unnamed: 2_level_0':'General','Unnamed: 3_level_0':'General'}, inplace=True)
team_stats_against
```

```python
# upload model for making predictions
path = Path('../../models/xgboost_predictor_v4_weightsAndBonus.pickle')
with open(path,'rb') as f:
    model = pickle.load(f)
```

```python
# pre-allocate
players21['goalkeeper'] = 0
players21['defender'] = 0
players21['midfielder'] = 0
players21['forward'] = 0
```

```python
input_columns = ['goalkeeper', 'defender', 'midfielder',
       'forward', 'xG90', 'xA90', 'xGA90', 'bonus', 'xPoints90',
       'games played', 'form 5', 'games_form5', 'form 10', 'games form10',
       'team xG per 90', 'team xGA per 90', 'home', 'opponent xG per 90',
       'opponent xGA per 90']
players21[input_columns] = numpy.nan
```

```python
# run predictions for all players for a given time horizon
# assumes one match per gameweek at most
horizon = 1
output_columns = [f'gameweek {i} prediction' for i in range(1,horizon+1)]
# THIS LOOKS TO BE IN THE WRONG PLACE?!?!
xGA_columns = [column for column in target_player_from20.columns if 'xGA_week' in column]

for player_ix in players21.index:
    target_player_from20 = players20.loc[(players20['first_name']==players21.loc[player_ix,'first_name']) & \
              (players20['second_name']==players21.loc[player_ix,'second_name'])]
    # if we find a unique match for current player from the 2020-21 data
    if target_player_from20.shape[0]==1:
        # collect base info
        if players21.loc[player_ix,'position'] == 'goalkeeper':
            players21.loc[player_ix,'goalkeeper'] = 1
        elif players21.loc[player_ix, 'position'] == 'defender':
            players21.loc[player_ix,'defender'] = 1
        elif players21.loc[player_ix, 'position'] == 'midfielder':
            players21.loc[player_ix,'midfielder'] = 1
        elif players21.loc[player_ix, 'position'] == 'forward':
            players21.loc[player_ix,'forward'] = 1
            
        players21.loc[player_ix, 'xG90'] = (90 * target_player_from20['xG'] / \
                                                 target_player_from20['minutes']).values[0]
        players21.loc[player_ix, 'xA90'] = (90 * target_player_from20['xA'] / \
                                                target_player_from20['minutes']).values[0]
        players21.loc[player_ix, 'xGA90'] = ( 90 * target_player_from20[xGA_columns].sum(axis=1) /
                                        target_player_from20['minutes'] ).values[0]
        players21.loc[player_ix, 'bonus'] = 0
        players21.loc[player_ix, 'xPoints90'] = target_player_from20['adjusted points per game'].values[0]
        players21.loc[player_ix, 'games played'] = 0
        players21.loc[player_ix, 'form 5'] = players21.loc[player_ix, 'xPoints90'].copy()
        players21.loc[player_ix, 'games_form5'] = 0
        players21.loc[player_ix, 'form 10'] = players21.loc[player_ix, 'xPoints90'].copy()
        players21.loc[player_ix, 'games form10'] = 0
        team = target_player_from20['team_name'].values[0]
        # team xG per 90 
        team_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==team, \
                                       ('Per 90 Minutes','xG')].values[0]
        players21.loc[player_ix, 'team xG per 90'] = team_xG90
        # team xGA per 90 
        team_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==f'vs {team}', \
                                                        ('Per 90 Minutes','xG')].values[0]
        players21.loc[player_ix, 'team xGA per 90'] = team_xGA90
        for gameweek in range(1,horizon+1):
            
            # home game
            if fpl_fixtures[(fpl_fixtures['event']==gameweek) & (fpl_fixtures['home team']==team)].shape[0]==1:
                players21.loc[player_ix, 'home'] = 1
                opponent = fpl_fixtures.loc[(fpl_fixtures['event']==gameweek) & (fpl_fixtures['home team']==team),\
                                            'away team'].values[0]
                if opponent in ['Brentford', 'Norwich City', 'Watford']:
                    # using Fulham's 20-21 stats here 
                    opponent_xG90 = 1.09
                    opponent_xGA90 = 1.39
                else:
                    # opponent xG per 90
                    opponent_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==opponent, \
                                                            ('Per 90 Minutes','xG')].values[0]
                    players21.loc[player_ix, 'opponent xG per 90'] = opponent_xG90
                    # opponent xGA per 90
                    opponent_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==\
                                                            f'vs {opponent}', ('Per 90 Minutes','xG')].values[0]
                    players21.loc[player_ix, 'opponent xGA per 90'] = opponent_xGA90
                
                # predict player's points
                input_data = players21.loc[player_ix:player_ix, input_columns]
                input_data = xgboost.DMatrix(input_data)
                prediction = model.predict(input_data)
                players21.loc[player_ix, output_columns[gameweek-1]] = prediction
            # away game
            elif fpl_fixtures[(fpl_fixtures['event']==gameweek) & (fpl_fixtures['away team']==team)].shape[0]==1:
                players21.loc[player_ix, 'home'] = 0
                opponent = fpl_fixtures.loc[(fpl_fixtures['event']==gameweek) & (fpl_fixtures['away team']==team),\
                                            'home team'].values[0]
                if opponent in ['Brentford', 'Norwich City', 'Watford']:
                    # using Fulham's 20-21 stats here 
                    opponent_xG90 = 1.09
                    opponent_xGA90 = 1.39
                else:
                    # opponent xG per 90
                    opponent_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==opponent, \
                                                            ('Per 90 Minutes','xG')].values[0]
                    players21.loc[player_ix, 'opponent xG per 90'] = opponent_xG90
                    # opponent xGA per 90
                    opponent_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==\
                                                            f'vs {opponent}', ('Per 90 Minutes','xG')].values[0]
                    players21.loc[player_ix, 'opponent xGA per 90'] = opponent_xGA90
                
                # predict player's points
                input_data = players21.loc[player_ix:player_ix, input_columns]
                input_data = xgboost.DMatrix(input_data)
                prediction = model.predict(input_data)
                players21.loc[player_ix, output_columns[gameweek-1]] = prediction
```

```python
players21[f'Total xPoints in {horizon} gameweeks'] = players21[output_columns].sum(axis=1)
```

```python
players21[f'adjusted points per game times horizon ({horizon})'] = players21['adjusted points per game'] * horizon
```
