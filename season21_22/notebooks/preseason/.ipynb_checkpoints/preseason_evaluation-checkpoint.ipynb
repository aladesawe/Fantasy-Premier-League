```python
# data for new season
fpl_data = json.loads(requests.get('https://fantasy.premierleague.com/api/bootstrap-static/').text)
players21 = pandas.DataFrame(fpl_data['elements'])
players21.set_index('id',inplace=True)
```

```python
for player_ix in players21.index:
    target_player_from20 = players20.loc[(players20['first_name']==players21.loc[player_ix,'first_name']) & \
              (players20['second_name']==players21.loc[player_ix,'second_name'])]
    if target_player_from20.shape[0]==1:
        players21.loc[player_ix, 'adjusted points per game'] = \
                        target_player_from20['adjusted points per game'].values[0]
```

```python
players21[players21['adjusted points per game'].isnull()].shape
```

```python
players21.loc[players21['adjusted points per game'].isnull(), ['first_name','second_name']]
```

```python
players21['now_cost'] = players21['now_cost'] / 10.0
players21['value'] = players21['adjusted points per game'] / players21['now_cost']
```

```python
players21[players21['minutes']>900].sort_values(by='adjusted points per game', ascending=False).head(20)\
            [['first_name','second_name','adjusted points per game','value']]
```

```python
players21[players21['minutes']>900].sort_values(by='value', ascending=False).head(20)\
            [['first_name','second_name','now_cost','adjusted points per game','value']]
```

```python
players21['valuePoints metric'] = numpy.sqrt(players21['adjusted points per game'] * players21['value'])
```

```python
players21[players21['minutes']>900].sort_values(by='valuePoints metric', ascending=False).head(20)\
            [['first_name','second_name','adjusted points per game','value','valuePoints metric']]
```
