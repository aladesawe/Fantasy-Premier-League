```python
%%time
rng = numpy.random.default_rng(42)

# create full data set
data = []
for player_ix in players.index:
    for column_ix in range(0,len(xPoints_columns)):
        if ~numpy.isnan(players.loc[player_ix,xPoints_columns[column_ix]]):

            # add player id for later easy access to a given player's predictions (not used in modelling)
            new_row = [player_ix]
            # add gameweek xPoints
            new_row = new_row + [players.loc[player_ix,xPoints_columns[column_ix]]]
            
            # player position with one-hot-encoding
            position_encoding = [0,0,0,0]
            element_type = players.loc[player_ix, 'element_type']
            position_encoding[element_type-1] = 1
            new_row = new_row + position_encoding
            
            # add underlying performance metrics (per 90 minutes) and in how many games the player has played 
            gameweek_appearances = players.loc[player_ix, xPoints_columns[:column_ix]].notnull().sum()
            gameweek_appearances_season = players.loc[player_ix, xPoints_columns].notnull().sum()
            if gameweek_appearances > 0:
                xG90 = players.loc[player_ix, xG_columns[:column_ix]].sum() / gameweek_appearances
                xA90 = players.loc[player_ix, xA_columns[:column_ix]].sum() / gameweek_appearances
                xGA90 = players.loc[player_ix, xGA_columns[:column_ix]].sum() / gameweek_appearances
                bonus = players.loc[player_ix,'bps'] * (gameweek_appearances / gameweek_appearances_season) 
                xPoints_90 = players.loc[player_ix, xPoints_columns[:column_ix]].sum() / gameweek_appearances
            else:
                xG90 = 0
                xA90 = 0
                xGA90 = 0
                bonus = 0
                xPoints_90 = 0
            
            new_row = new_row + [xG90, xA90, xGA90, bonus, xPoints_90, gameweek_appearances]
            
            # calculate player form for past 5 gameweeks
            if column_ix > 4:
                form_columns = xPoints_columns[column_ix-5:column_ix]
            else:
                form_columns = xPoints_columns[:column_ix]
            form5 = players.loc[player_ix, form_columns].sum()
            gameweek_appearances_form5 = players.loc[player_ix, form_columns].notnull().sum()
            new_row = new_row + [form5, gameweek_appearances_form5]
            
            # calculate player form for past 10 gameweeks
            if column_ix > 9:
                form_columns = xPoints_columns[column_ix-10:column_ix]
            else:
                form_columns = xPoints_columns[:column_ix]
            form10 = players.loc[player_ix, form_columns].sum()
            gameweek_appearances_form10 = players.loc[player_ix, form_columns].notnull().sum()
            new_row = new_row + [form10, gameweek_appearances_form10]
            
            team = players.loc[player_ix,'team_name']
            # team xG per 90 + random noise (random noise so that training does not pick up too closely on 
            # specific teams)
            team_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==team, \
                                           ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [rng.normal(team_xG90, 0.1)]
            # team xGA per 90
            team_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==f'vs {team}', \
                                                        ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [rng.normal(team_xGA90, 0.1)]
            # check whether the game is home or away and add team and opponent characteristics
            # for now, we disregard player data from double (or triple!) gameweeks
            if fixtures[(fixtures['Week']==column_ix+1) & ((fixtures['Home']==team) | (fixtures['Away']==team))].shape[0]==1:
                if fixtures[(fixtures['Week']==column_ix+1) & (fixtures['Home']==team)].shape[0]==1:
                    # home game
                    new_row = new_row + [1]
                    # opponent 
                    opponent = fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Home']==team), 'Away'].values[0]
                elif fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Away']==team)].shape[0]==1:
                    # away game
                    new_row = new_row + [0]
                    # opponent 
                    opponent = fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Away']==team), 'Home'].values[0]
            else:
                continue;
                
            # opponent xG per 90
            opponent_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==opponent, \
                                                        ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [rng.normal(opponent_xG90, 0.1)]
            # opponent xGA per 90
            opponent_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==\
                                                        f'vs {opponent}', ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [rng.normal(opponent_xGA90, 0.1)]
            
            data.append(new_row)

# 'xPoints per game', 'form 5', 'form 10'
data = pandas.DataFrame(data, columns=['player_index','xPoints', 'goalkeeper', 'defender', 'midfielder', 'forward', 'xG90', \
                                       'xA90', 'xGA90', 'bonus', 'xPoints90', 'games played', 'form 5', 'games_form5', \
                                       'form 10', 'games form10', 'team xG per 90', 'team xGA per 90', 'home', \
                                       'opponent xG per 90', 'opponent xGA per 90'])
print(f'Data shape: {data.shape}')
```

# Train model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error
import xgboost
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
data.columns
```

```python
# target
y = data['xPoints']
# input
x = data[data.columns[2:]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
sample_weights = 0.5*numpy.ones(x_train.shape[0])
sample_weights[(x_train['xPoints90']>=4.5)] = 1
sample_weights
```

```python
# change training data into Dmatrix-form
#dtrain = xgboost.DMatrix(x_train, label=y_train)
dtrain = xgboost.DMatrix(x_train, label=y_train, weight=sample_weights)
dtest = xgboost.DMatrix(x_test, label=y_test)
```

```python
params = {
    # Parameters that we are going to tune.
    'max_depth':6, 'min_child_weight': 1, 'eta':.3,
    'subsample': 1, 'colsample_bytree': 1,
    # Other parameters 
    'objective':'reg:squarederror',
}
```

### Check for optimal number of boosting rounds in the otherwise base model

```python
params['eval_metric'] = "rmse"
```

```python
num_boost_round = 999
```

```python
model = xgboost.train(
params,
dtrain, 
num_boost_round=num_boost_round, 
evals=[(dtest, "Test")], 
early_stopping_rounds=10
)
```

```python
print("Best RMSE: {:.2f} with {} rounds".format( model.best_score,
model.best_iteration+1))
```

### Hyper-paremeter tuning with cross-validation

```python
cv_results = xgboost.cv(
    params,
    dtrain,
    num_boost_round=num_boost_round, 
    seed=42,
    nfold=5,
    metrics={'rmse'},
    early_stopping_rounds=10
) 

cv_results
```

### Optimize max_depth and min_child_weight

```python
gridsearch_params = [
(max_depth, min_child_weight)
for max_depth in range(4,13,2)
for min_child_weight in numpy.arange(1,16,2)
]
print(len(gridsearch_params))
print(gridsearch_params)
```

```python
params
```

```python
# Define initial best params and RMSE
min_rmse = float("Inf")
best_params = None
for max_depth, min_child_weight in gridsearch_params:
    print(f"CV with max_depth={max_depth}, min_child_weight={min_child_weight}")
    # Update our parameters
    params['max_depth'] = max_depth 
    params['min_child_weight'] = min_child_weight
    # Run CV
    cv_results = xgboost.cv( 
        params,
        dtrain, 
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'rmse'},
        early_stopping_rounds=10
    )
    # Update best MAE
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin() 
    print(f"\tRMSE {mean_rmse} for {boost_rounds} rounds") 
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = (max_depth,min_child_weight)
        
print(f"Best params: {best_params[0]}, {best_params[1]}, RMSE: {min_rmse}")
```

```python
gridsearch_params = [
    (max_depth, min_child_weight)
    for max_depth in range(1,6)
    for min_child_weight in range(1,8)
]
print(len(gridsearch_params))
print(gridsearch_params)
```

```python
# Define initial best params and RMSE
min_rmse = float("Inf")
best_params = None
for max_depth, min_child_weight in gridsearch_params:
    print(f"CV with max_depth={max_depth}, min_child_weight={min_child_weight}")
    # Update our parameters
    params['max_depth'] = max_depth 
    params['min_child_weight'] = min_child_weight
    # Run CV
    cv_results = xgboost.cv( 
        params,
        dtrain, 
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'rmse'},
        early_stopping_rounds=10
    )
    # Update best MAE
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin() 
    print(f"\tRMSE {mean_rmse} for {boost_rounds} rounds") 
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = (max_depth,min_child_weight)
        
print(f"Best params: {best_params[0]}, {best_params[1]}, RMSE: {min_rmse}")
```

```python
# update best params
params['max_depth'] = 4
params['min_child_weight'] = 3
```

### Optimize subsample and colsample

```python
gridsearch_params = [
    (subsample, colsample)
    for subsample in [i/10. for i in range(7,11)] 
    for colsample in [i/10. for i in range(4,11)]
]
gridsearch_params
```

```python
min_rmse = float("Inf") 
best_params = None
# We start by the largest values and go down to the smallest
for subsample, colsample in reversed(gridsearch_params): 
    print("CV with subsample={}, colsample={}".format(subsample, colsample))
    # We update our parameters
    params['subsample'] = subsample 
    params['colsample_bytree'] = colsample
    # Run CV
    cv_results = xgboost.cv( 
        params,
        dtrain, 
        num_boost_round=num_boost_round, 
        seed=42,
        nfold=5,
        metrics={'rmse'}, early_stopping_rounds=10
    )
    # Update best score
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin() 
    print("\tRMSE {} for {} rounds".format(mean_rmse, boost_rounds)) 
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = (subsample,colsample)

print("Best params: {}, {}, RMSE: {}".format(best_params[0], best_params[1], min_rmse))
```

```python
params['subsample'] = 1.
params['colsample_bytree'] = 1.
```

### Optimize eta

```python
%%time

min_rmse = float("Inf") 
best_params = None
for eta in [0.5, 0.4, .3, .2, .1, .05, .01, .005]: 
    print("CV with eta={}".format(eta))
    # We update our parameters
    params['eta'] = eta
    # Run and time CV
    %time 
    cv_results = xgboost.cv( 
        params,
        dtrain, 
        num_boost_round=num_boost_round, 
        seed=42,
        nfold=5,
        metrics=['rmse'],
        early_stopping_rounds=10
    )
    # Update best score
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin() 
    print("\tRMSE {} for {} rounds\n".format(mean_rmse, boost_rounds)) 
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse 
        best_params = eta

print("Best params: {}, RMSE: {}".format(best_params, min_rmse))
```

# Comparison to a "naive" model

The naive model here is simply predicting the player's mean points per game up to gameweek at hand. 

```python
predictions = []
actuals = []
for player_ix in players.index:
    for column_ix in range(0,len(xPoints_columns)):
        if ~numpy.isnan(players.loc[player_ix,xPoints_columns[column_ix]]):
            # calculate xPoints per game up to current point in the season
            mean_Xpoints_columns = xPoints_columns[:column_ix] 
            # don't add row if no xPoints per game can be calculated
            if numpy.isnan(players.loc[player_ix,mean_Xpoints_columns].mean()):
                continue;
            predictions = predictions + [players.loc[player_ix,mean_Xpoints_columns].mean()]
            actuals = actuals + [players.loc[player_ix,xPoints_columns[column_ix]]]

print(len(predictions))
print(len(actuals))
```

```python
mse = mean_squared_error(actuals, predictions)
print(f'Explained variance: {explained_variance_score(actuals, predictions)}')
print(f'MSE: {mean_squared_error(actuals, predictions)}')
print(f'MAE: {mean_absolute_error(actuals, predictions)}')
```

```python
if mean_squared_error(actuals, predictions)-mean_squared_error(y_test, y_pred)>0:
    print(f'XGBoost improves MSE from "naive" prediction by {100*(mean_squared_error(actuals, predictions)-mean_squared_error(y_test, y_pred))/mean_squared_error(actuals, predictions):.2f} percent.')
else:
    print(f'XGBoost is worse (in MSE) than "naive" prediction by {100*(mean_squared_error(y_test, y_pred)-mean_squared_error(actuals, predictions))/mean_squared_error(actuals, predictions):.2f} percent.')
    
if explained_variance_score(y_test, y_pred) - explained_variance_score(actuals, predictions) > 0:
    print(f'XGBoost improves explained variance by {100*(explained_variance_score(y_test, y_pred) - explained_variance_score(actuals, predictions)):.2f} percent.')
else:
    print(f'XGBoost is worse in explained variance by {100*(explained_variance_score(actuals, predictions) - explained_variance_score(y_test, y_pred)):.2f} percent.')
```
