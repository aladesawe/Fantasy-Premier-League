```python
%%time

# create full data set
# collect variables: ['xPoints', 'xG per 90', 'xA per 90', 'xGA per 90', 'position', 'form 5', 'form 10', 
#                    'team xG per 90', 'team xGA per 90', 'home', 'opponent xG per 90', 'opponent xGA per 90']

data = []

for player_ix in players.index:
    for column_ix in range(0,len(xPoints_columns)):
        if ~numpy.isnan(players.loc[player_ix,xPoints_columns[column_ix]]):

            # add player id for later easy access to a given player's predictions (not used in modelling)
            new_row = [player_ix]
            # add gameweek xPoints
            new_row = new_row + [players.loc[player_ix,xPoints_columns[column_ix]]]
            
            # player position with one-hot-encoding
            position_encoding = [0,0,0,0]
            element_type = players.loc[player_ix, 'element_type']
            position_encoding[element_type-1] = 1
            new_row = new_row + position_encoding
            
            # add underlying performance metrics (per 90 minutes) and in how many games the player has played 
            gameweek_appearances = players.loc[player_ix, xPoints_columns[:column_ix]].notnull().sum()
            if gameweek_appearances > 0:
                xG90 = players.loc[player_ix, xG_columns[:column_ix]].sum() / gameweek_appearances
                xA90 = players.loc[player_ix, xA_columns[:column_ix]].sum() / gameweek_appearances
                xGA90 = players.loc[player_ix, xGA_columns[:column_ix]].sum() / gameweek_appearances
                xPoints_90 = players.loc[player_ix, xPoints_columns[:column_ix]].sum() / gameweek_appearances
            else:
                xG90 = 0
                xA90 = 0
                xGA90 = 0
                xPoints_90 = 0
            
            new_row = new_row + [xG90, xA90, xGA90, xPoints_90, gameweek_appearances]
            
            # calculate player form for past 5 gameweeks
            if column_ix > 4:
                form_columns = xPoints_columns[column_ix-5:column_ix]
            else:
                form_columns = xPoints_columns[:column_ix]
            form5 = players.loc[player_ix, form_columns].sum()
            gameweek_appearances_form5 = players.loc[player_ix, form_columns].notnull().sum()
            new_row = new_row + [form5, gameweek_appearances_form5]
            
            # calculate player form for past 10 gameweeks
            if column_ix > 9:
                form_columns = xPoints_columns[column_ix-10:column_ix]
            else:
                form_columns = xPoints_columns[:column_ix]
            form10 = players.loc[player_ix, form_columns].sum()
            gameweek_appearances_form10 = players.loc[player_ix, form_columns].notnull().sum()
            new_row = new_row + [form10, gameweek_appearances_form10]
            
            team = players.loc[player_ix,'team_name']
            # team xG per 90 + random noise (random noise so that training does not pick up too closely on 
            # specific teams)
            team_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==team, \
                                           ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [numpy.random.normal(team_xG90, 0.1)]
            # team xGA per 90
            team_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==f'vs {team}', \
                                                        ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [numpy.random.normal(team_xGA90, 0.1)]
            # check whether the game is home or away and add team and opponent characteristics
            # for now, we disregard player data from double (or triple!) gameweeks
            if fixtures[(fixtures['Week']==column_ix+1) & ((fixtures['Home']==team) | (fixtures['Away']==team))].shape[0]==1:
                if fixtures[(fixtures['Week']==column_ix+1) & (fixtures['Home']==team)].shape[0]==1:
                    # home game
                    new_row = new_row + [1]
                    # opponent 
                    opponent = fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Home']==team), 'Away'].values[0]
                elif fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Away']==team)].shape[0]==1:
                    # away game
                    new_row = new_row + [0]
                    # opponent 
                    opponent = fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Away']==team), 'Home'].values[0]
            else:
                continue;
                
            # opponent xG per 90
            opponent_xG90 = team_stats_for.loc[team_stats_for[('General','Squad')]==opponent, \
                                                        ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [numpy.random.normal(opponent_xG90, 0.1)]
            # opponent xGA per 90
            opponent_xGA90 = team_stats_against.loc[team_stats_against[('General','Squad')]==\
                                                        f'vs {opponent}', ('Per 90 Minutes','xG')].values[0]
            new_row = new_row + [numpy.random.normal(opponent_xGA90, 0.1)]
            
            data.append(new_row)

# 'xPoints per game', 'form 5', 'form 10'
data = pandas.DataFrame(data, columns=['player_index','xPoints', 'goalkeeper', 'defender', 'midfielder', 'forward', 'xG90', \
                                       'xA90', 'xGA90', 'xPoints90', 'games played', 'form 5', 'games_form5', \
                                       'form 10', 'games form10', 'team xG per 90', 'team xGA per 90', 'home', \
                                       'opponent xG per 90', 'opponent xGA per 90'])
print(f'Data shape: {data.shape}')
```

# Train model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
# target
y = data['xPoints']
# input
x = data[data.columns[2:]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model = XGBRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(f'Explained variance: {explained_variance_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {numpy.sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
```

# Comparison to a "naive" model

The naive model here is simply predicting the player's mean points per game up to gameweek at hand. 

```python
predictions = []
actuals = []
for player_ix in players.index:
    for column_ix in range(0,len(xPoints_columns)):
        if ~numpy.isnan(players.loc[player_ix,xPoints_columns[column_ix]]):
            # calculate xPoints per game up to current point in the season
            mean_Xpoints_columns = xPoints_columns[:column_ix] 
            # don't add row if no xPoints per game can be calculated
            if numpy.isnan(players.loc[player_ix,mean_Xpoints_columns].mean()):
                continue;
            predictions = predictions + [players.loc[player_ix,mean_Xpoints_columns].mean()]
            actuals = actuals + [players.loc[player_ix,xPoints_columns[column_ix]]]

print(len(predictions))
print(len(actuals))
```

```python
mse = mean_squared_error(actuals, predictions)
print(f'Explained variance: {explained_variance_score(actuals, predictions)}')
print(f'MSE: {mean_squared_error(actuals, predictions)}')
print(f'MAE: {mean_absolute_error(actuals, predictions)}')
```

```python
if mean_squared_error(actuals, predictions)-mean_squared_error(y_test, y_pred)>0:
    print(f'XGBoost improves from "naive" prediction by {100*(mean_squared_error(actuals, predictions)-mean_squared_error(y_test, y_pred))/mean_squared_error(actuals, predictions):.2f} percent.')
else:
    print(f'XGBoost is worse than "naive" prediction by {100*(mean_squared_error(y_test, y_pred)-mean_squared_error(actuals, predictions))/mean_squared_error(actuals, predictions):.2f} percent.')
```
