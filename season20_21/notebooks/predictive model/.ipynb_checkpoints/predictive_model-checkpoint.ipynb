Here we want to build a simple model for predicting a team's xG for a given match, taking into account match location (home or away) and opposition quality (i.e., mean xGA of the opponent). 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import statsmodels.api as sm
from scipy.stats import poisson
pd.set_option("display.max_columns", 50)
```

```python
# import team data and pre-process
filepath = '../../data/fbref/team_stats_season19_20.csv'
teamStats = pd.read_csv(filepath, index_col='Squad')
teamStats.head()
```

```python
filepath = '../../data/fbref/match_results_season19_20.csv'
fixtures = pd.read_csv(filepath)
fixtures.rename(columns={'Wk':'Gameweek', 'xG':'xG_home', 'xG.1':'xG_away'}, inplace=True)
fixtures.head()
```

Simple linear regression looking at xG relative to opponent mean xGA (over the whole season).

```python
all_xG = []
all_opp_xGA = []
for team in teamStats.index:
    # collect target team xGs
    xG_home = fixtures.loc[fixtures['Home']==team, 'xG_home'].values
    xG_away = fixtures.loc[fixtures['Away']==team, 'xG_away'].values
    xG = np.concatenate((xG_home,xG_away))
    all_xG.append(xG)
    # collect target team opponents' xGAs
    home_opp = fixtures.loc[fixtures['Home']==team, 'Away'].values
    away_opp = fixtures.loc[fixtures['Away']==team, 'Home'].values
    home_opp_xGA = teamStats.loc[home_opp, 'xGA per game'].values
    away_opp_xGA = teamStats.loc[away_opp, 'xGA per game'].values
    opp_xGA = np.concatenate((home_opp_xGA, away_opp_xGA))
    all_opp_xGA.append(opp_xGA)
```

```python
# REGRESSION WITHOUT BIAS TERM

# perform linear regression for each team's xG 
all_results = []
count = 0
error_count = 0
for team in teamStats.index:
    print(team)
    X = all_opp_xGA[count]
    Y = all_xG[count]
    model = sm.OLS(Y, X)
    results = model.fit()
    if results.pvalues[0]>=0.05:
        print(team + ': regression coefficient not significant.')
        error_count += 1    
    print(results.summary())
    print()
    all_results.append(results.params)
    count += 1

if error_count == 0:
    print()
    print('Regression was found for each team.')
else:
    print()
    print('Not all regressions were found.')
```

```python
regression_coefficients = pd.DataFrame(all_results, index=teamStats.index, columns=['coefficient'])
display(regression_coefficients)
```

Linear regression for the coefficients.

```python
X = teamStats['xG per game'].values
Y = regression_coefficients.values
model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
coef = results.params
x = np.linspace(0,3,100)
plt.scatter(X,Y)
plt.plot(x,coef*x,'black')
plt.xlabel('xG')
plt.ylabel('regression coefficient')
plt.show()
```

Use this to define a team's xG regression coefficient.

```python
print('xG_regression_coefficient: ' + str(coef[0]))
```

Calculate home and away coefficients for xG.

```python
homexG = fixtures['xG_home'].mean()
awayxG = fixtures['xG_away'].mean()
teamxG = teamStats['xG per game'].mean()
home_coefficient = homexG / teamxG
away_coefficient = awayxG / teamxG
print('Average home team xG: ' + str(homexG))
print('Average away team xG: ' + str(awayxG))
print('Average team xG: ' + str(teamxG))
print('Home xG coefficient: ' + str(home_coefficient))
print('Away xG coefficient: ' + str(away_coefficient))
```

```python
x = np.linspace(0,2,100)
count=0
for team in teamStats.index:
    X = all_opp_xGA[count]
    Y = all_xG[count]
    plt.scatter(X,Y)
    plt.plot(x,all_results[count]*x,'black')
    plt.plot(x,home_coefficient*all_results[count]*x,'red')
    plt.plot(x,away_coefficient*all_results[count]*x,'blue')
    plt.xlabel('opponent xGA_per_game')
    plt.ylabel('xG')
    plt.title(team)
    plt.legend(('mean','home','away'))
    plt.xlim([0.8,2])
    plt.show()
    count += 1
```
