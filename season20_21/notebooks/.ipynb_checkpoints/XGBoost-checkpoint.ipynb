```python
players_over900minutes = players[players['minutes']>900]
players_over900minutes
```

```python
player_profiles = pandas.DataFrame(index=players_over900minutes.index)
player_profiles['xG per 90'] = 90 * players_over900minutes['xG'] / players_over900minutes['minutes']
player_profiles['xA per 90'] = 90 * players_over900minutes['xA'] / players_over900minutes['minutes']
player_profiles
```

```python
xGA_columns = [column for column in players_over900minutes.columns if 'xGA_week' in column]
player_profiles['xGA per 90'] = ( 90 * players_over900minutes[xGA_columns].sum(axis=1) /
                                        players_over900minutes['minutes'] )
player_profiles
```

```python
player_profiles['position'] = players_over900minutes['element_type']
player_profiles
```

## CREATE DATA SET

```python
# get relevant column names
xPoints_columns = ['xPoints week 1', 'xPoints week 2', 'xPoints week 3', 'xPoints week 4',
 'xPoints week 5', 'xPoints week 6', 'xPoints week 7', 'xPoints week 8', 'xPoints week 9',
 'xPoints week 10', 'xPoints week 11', 'xPoints week 12', 'xPoints week 13', 'xPoints week 14',
 'xPoints week 15', 'xPoints week 16', 'xPoints week 17', 'xPoints week 18', 'xPoints week 19',
 'xPoints week 20', 'xPoints week 21', 'xPoints week 22', 'xPoints week 23', 'xPoints week 24',
 'xPoints week 25', 'xPoints week 26', 'xPoints week 27', 'xPoints week 28', 'xPoints week 29',
 'xPoints week 30', 'xPoints week 31', 'xPoints week 32', 'xPoints week 33', 'xPoints week 34',
 'xPoints week 35', 'xPoints week 36', 'xPoints week 37', 'xPoints week 38']

xPoints_columns
```

```python
players_over900minutes[xPoints_columns].dtypes
```

```python
players_over900minutes.loc[2,xPoints_columns]
```

```python
%%time

# create full data set
# collect variables: ['xPoints', 'xG per 90', 'xA per 90', 'xGA per 90', 'position', 'form 5', 'form 10', 
#                    'team xG per 90', 'team xGA per 90', 'home', 'opponent xG per 90', 'opponent xGA per 90']

data = []

for player_ix in players_over900minutes.index:
    for column_ix in range(0,len(xPoints_columns)):
        if ~numpy.isnan(players_over900minutes.loc[player_ix,xPoints_columns[column_ix]]):

            # basic player attributes
            new_row = [
                players_over900minutes.loc[player_ix,xPoints_columns[column_ix]],
                player_profiles.loc[player_ix,'xG per 90'],
                player_profiles.loc[player_ix,'xA per 90'],
                player_profiles.loc[player_ix,'xGA per 90'],                
                  ]
            # player position with one-hot-encoding
            position = player_profiles.loc[player_ix,'position']
            position_encoding = [0,0,0,0]
            position_encoding[position-1] = 1
            new_row = new_row + position_encoding
            
            # add player's mean points per game over the whole season (this needs to be evaluated 
            # somehow "in production", e.g., with a Bayesian framework)
            # estimate of a player's general FPL production level
            # total xPoints
            #xPoints_per_game = players_over900minutes.loc[player_ix, xPoints_columns].sum() / \
            #                    players_over900minutes.loc[player_ix,xPoints_columns].notnull().sum()
            #new_row = new_row + [xPoints_per_game]
            
            # calculate player form for past 5 gameweeks
            if column_ix > 4:
                form_columns = xPoints_columns[column_ix-5:column_ix]
            else:
                form_columns = xPoints_columns[:column_ix]
            # don't add row if no form can be calculated
            if numpy.isnan(players_over900minutes.loc[player_ix,form_columns].mean()):
                continue;
            new_row = new_row + [players_over900minutes.loc[player_ix,form_columns].mean()]
            
            # calculate player form for past 10 gameweeks
            if column_ix > 9:
                form_columns = xPoints_columns[column_ix-10:column_ix]
            else:
                form_columns = xPoints_columns[:column_ix]
            # don't add row if no form can be calculated
            if numpy.isnan(players_over900minutes.loc[player_ix,form_columns].mean()):
                continue;
            new_row = new_row + [players_over900minutes.loc[player_ix,form_columns].mean()]
            
            # calculate xPoints per game up to current point in the season
            mean_Xpoints_columns = xPoints_columns[:column_ix] 
            # don't add row if no xPoints per game can be calculated
            if numpy.isnan(players_over900minutes.loc[player_ix,mean_Xpoints_columns].mean()):
                continue;
            new_row = new_row + [players_over900minutes.loc[player_ix,mean_Xpoints_columns].mean()]
            # add info on which game week it is (migth be useful for the model because mean points per game
            # fluctuates early season and stabilizes towards the end)
            new_row = new_row + [column_ix+1]
            
            team = players_over900minutes.loc[player_ix,'team_name']
            # team xG per 90
            new_row = new_row + [team_stats_for.loc[team_stats_for[('General','Squad')]==team, \
                                                        ('Per 90 Minutes','xG')].values[0]]
            # team xGA per 90
            new_row = new_row + [team_stats_against.loc[team_stats_against[('General','Squad')]==f'vs {team}', \
                                                        ('Per 90 Minutes','xG')].values[0]]
            # check whether the game is home or away and add team and opponent characteristics
            # for now, we disregard player data from double (or triple!) gameweeks
            if fixtures[(fixtures['Week']==column_ix+1) & ((fixtures['Home']==team) | (fixtures['Away']==team))].shape[0]==1:
                if fixtures[(fixtures['Week']==column_ix+1) & (fixtures['Home']==team)].shape[0]==1:
                    # home game
                    new_row = new_row + [1]
                    # opponent 
                    opponent = fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Home']==team), 'Away'].values[0]
                elif fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Away']==team)].shape[0]==1:
                    # away game
                    new_row = new_row + [0]
                    # opponent 
                    opponent = fixtures.loc[(fixtures['Week']==column_ix+1) & (fixtures['Away']==team), 'Home'].values[0]
            else:
                continue;
                
            # opponent xG per 90
            new_row = new_row + [team_stats_for.loc[team_stats_for[('General','Squad')]==opponent, \
                                                        ('Per 90 Minutes','xG')].values[0]]
            # opponent xGA per 90
            new_row = new_row + [team_stats_against.loc[team_stats_against[('General','Squad')]==\
                                                        f'vs {opponent}', ('Per 90 Minutes','xG')].values[0]]
            
            data.append(new_row)

# 'xPoints per game', 'form 5', 'form 10'
data = pandas.DataFrame(data, columns=['xPoints', 'xG per 90', 'xA per 90', 'xGA per 90', 'goalkeeper', \
                                          'defender', 'midfielder', 'forward', 'form 5', 'form 10', \
                                       'xPoints per game', 'gameweek', 'team xG per 90', 'team xGA per 90', 'home', \
                                       'opponent xG per 90', 'opponent xGA per 90'])
print(f'Data shape: {data.shape}')
```

# Train model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
# target
y = data['xPoints']
# input
x = data[data.columns[1:]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model = XGBRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(f'Explained variance: {explained_variance_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
```

```python
plt.hist(y_pred,bins=numpy.arange(0,12))
plt.show()
```

```python
plt.hist(y_test-y_pred, bins=numpy.arange(-8,8))
plt.show()
```

```python
plt.scatter(y_test, y_pred)
diagonal_line=numpy.linspace(-3,16,100)
plt.plot(diagonal_line,diagonal_line,'r')
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()
```

# Comparison to a "naive" model

The naive model here is simply predicting the player's mean points per game up to gameweek at hand. 

```python
predictions = []
actuals = []
for player_ix in players_over900minutes.index:
    for column_ix in range(0,len(xPoints_columns)):
        if ~numpy.isnan(players_over900minutes.loc[player_ix,xPoints_columns[column_ix]]):
            # calculate xPoints per game up to current point in the season
            mean_Xpoints_columns = xPoints_columns[:column_ix] 
            # don't add row if no xPoints per game can be calculated
            if numpy.isnan(players_over900minutes.loc[player_ix,mean_Xpoints_columns].mean()):
                continue;
            predictions = predictions + [players_over900minutes.loc[player_ix,mean_Xpoints_columns].mean()]
            actuals = actuals + [players_over900minutes.loc[player_ix,xPoints_columns[column_ix]]]

print(len(predictions))
print(len(actuals))
```

```python
mse = mean_squared_error(actuals, predictions)
print(f'Explained variance: {explained_variance_score(actuals, predictions)}')
print(f'MSE: {mean_squared_error(actuals, predictions)}')
print(f'MAE: {mean_absolute_error(actuals, predictions)}')
```

```python
if mean_squared_error(actuals, predictions)-mean_squared_error(y_test, y_pred)>0:
    print(f'XGBoost improves from "naive" prediction by {100*(mean_squared_error(actuals, predictions)-mean_squared_error(y_test, y_pred))/mean_squared_error(actuals, predictions):.2f} percent.')
else:
    print(f'XGBoost is worse than "naive" prediction by {100*(mean_squared_error(y_test, y_pred)-mean_squared_error(actuals, predictions))/mean_squared_error(actuals, predictions):.2f} percent.')
```
