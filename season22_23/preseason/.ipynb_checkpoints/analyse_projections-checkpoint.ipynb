```python
import numpy as np
import pandas as pd
import json
import requests

import matplotlib.pyplot as plt
import plotly.graph_objects as go

# allow more data columns/rows to be shown than by default
pd.set_option('display.max_columns', 501)
pd.set_option('display.max_rows', 501)
```

```python
fpl_data = json.loads(requests.get('https://fantasy.premierleague.com/api/bootstrap-static/').text)
fpl_df = pd.DataFrame(fpl_data['elements'])
fpl_df.set_index('id',inplace=True)
fpl_df['name'] = fpl_df['first_name'] + ' ' + fpl_df['second_name']
fpl_df['name'] = fpl_df['name'].str.normalize('NFD').str.encode('ascii','ignore').str.decode('ascii')
fpl_df['now_cost'] = fpl_df['now_cost'] / 10.0
fpl_df
```

```python
projections = pd.read_csv('../data/pre_season_projections_full.csv', index_col=0)
projections
```

```python
# top players for each gameweek
for gameweek in range(1,11):
    print(f'Gameweek {gameweek}:')
    display(projections.loc[projections['gameweek']==gameweek].sort_values(by='xP', ascending=False).head(30))
    print()
```

```python
# calculate cumulative xP for each player
projections.groupby(by=['name']).sum().sort_values(by='xP', ascending=False)['xP'].head(30)
```

```python
# top players in points-per-game for different gameweek horizons
for gameweek in range(1,11):
    print(f'Horizon {gameweek}:')
    display(projections[projections['gameweek']<=gameweek].groupby(by=['name']).sum().\
            sort_values(by='xP', ascending=False)['xP'].head(20))
    print()
```

```python
projections[projections['name'].str.contains('Cash')]
```

```python
fig = go.Figure()

#players = ['Erling Haland', 'Harry Kane']
#players = ['Luis Diaz', 'Mason Mount', 'Gabriel Martinelli Silva']
#players = ['Oleksandr Zinchenko']
#players = ['Alisson Ramses Becker', 'Ederson Santana de Moraes', 'Edouard Mendy', 'David Raya Martin', 'Aaron Ramsdale']
#players = ['Ivan Toney', 'Gabriel Fernando de Jesus']
players = ['Matty Cash', 'Kieran Trippier']

for player in players:
    fig.add_trace(
        go.Scatter(
            x=projections.loc[projections['name']==player, 'gameweek'],
            y=projections.loc[projections['name']==player, 'xP'],
            mode="lines+markers",
            name=player,
            hovertext=('Opponent: ' + projections.loc[projections['name']==player, 'opp_team'] + 
                        ', was_home: ' + projections.loc[projections['name']==player, 'was_home'].astype(str)),
            showlegend=True,
            ),
    )

fig.update_layout(
    #title="",
    template='plotly_dark',
    xaxis_title="gameweek",
    yaxis_title='xP',
    #showlegend=True
)
```

```python
fig = go.Figure()

#players = ['Erling Haland', 'Harry Kane']
#players = ['Luis Diaz', 'Mason Mount', 'Gabriel Martinelli Silva']
#players = ['Ivan Toney', 'Gabriel Fernando de Jesus']
#players = ['Alisson Ramses Becker', 'Ederson Santana de Moraes', 'Edouard Mendy', 'David Raya Martin', 'Aaron Ramsdale']
players = ['Matty Cash', 'Kieran Trippier']

for player in players:
    fig.add_trace(
        go.Scatter(
            x=projections.loc[projections['name']==player, 'gameweek'],
            y=projections.loc[projections['name']==player, 'xP'].cumsum(),
            mode="lines+markers",
            name=player,
            #marker_color='light_blue',
            showlegend=True,
            ),
    )

fig.update_layout(
    #title="",
    template='plotly_dark',
    xaxis_title="gameweek",
    yaxis_title='cumulative xP',
    #showlegend=True
)
```

```python
# calculate ppg over 10 game horizon and geometric mean of ppg and cost
xPpg_10 = projections[projections['gameweek']<=11].groupby(by=['name']).sum()['xP'] / 10
xPpg_10 = xPpg_10.to_frame()
xPpg_10 = xPpg_10.reset_index()
xPpg_10 = xPpg_10.rename(columns={'xP':'xPpg_10'})
xPpg_10['price'] = [fpl_df.loc[fpl_df['name']==name, 'now_cost'].values[0] for name in xPpg_10['name']]
xPpg_10['value'] = xPpg_10['xPpg_10'] / xPpg_10['price']
xPpg_10['geometric_valuePoints'] = np.sqrt(xPpg_10['xPpg_10'] * xPpg_10['value'])
xPpg_10['position'] = [fpl_df.loc[fpl_df['name']==name, 'element_type'].values[0] for name in xPpg_10['name']]
position_dict={1:'GK', 2:'DEF', 3:'MID', 4:'FWD'}
xPpg_10['position'] = xPpg_10['position'].map(position_dict)
xPpg_10['team'] = [projections.loc[projections['name']==name, 'team'].head(1).values[0] for name in xPpg_10['name']]
xPpg_10
```

```python
session = requests.session()
```

```python
filepath = '../../../access_control/FPL_access.txt'
with open(filepath) as file:
    text = file.read().splitlines()
    email = text[0]
    password = text[1]
```

```python
url = 'https://users.premierleague.com/accounts/login/'

payload = {
'password': password,
'login': email,
'redirect_uri': 'https://fantasy.premierleague.com/a/login', 
#"redirect_uri": "https://fantasy.premierleague.com/",
#'redirect_uri': 'https://fantasy.premierleague.com/api/bootstrap-static/',
'app': 'plfpl-web'
}

headers = {
   'authority': 'users.premierleague.com' ,
   'cache-control': 'max-age=0' ,
   'upgrade-insecure-requests': '1' ,
   'origin': 'https://fantasy.premierleague.com' ,
   'content-type': 'application/x-www-form-urlencoded' ,
   'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36' ,
   'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9' ,
   'sec-fetch-site': 'same-site' ,
   'sec-fetch-mode': 'navigate' ,
   'sec-fetch-user': '?1' ,
   'sec-fetch-dest': 'document' ,
   'referer': 'https://fantasy.premierleague.com/my-team' ,
   'accept-language': 'en-US,en;q=0.9,he;q=0.8' ,
}

session.post(url, data=payload, headers=headers)
```

```python
response = session.get('https://fantasy.premierleague.com/api/my-team/243790')
team = response.json()['picks']
```

```python
my_player_ix = [player['element'] for player in team]
my_starting11 = my_player_ix[:-4]
my_subs = my_player_ix[-4:]
```

```python
projections['position'] = [fpl_df.loc[fpl_df['name']==name, 'element_type'].values[0] for name in projections['name']]
position_dict={1:'GK', 2:'DEF', 3:'MID', 4:'FWD'}
projections['position'] = projections['position'].map(position_dict)
projections['position'] = pd.Categorical(projections['position'], ['GK','DEF','MID','FWD'])
projections
```

```python
gameweek=1
expected_points = projections.loc[(projections['name'].isin(fpl_df.loc[my_starting11, 'name'].values)) & 
                (projections['gameweek']==gameweek), 'xP']
total_points = expected_points.sum() + expected_points.max()
print(f'Gameweek {gameweek}')
print(f'Projected points: {total_points}')
print()
print('Starting 11:')
display(projections.loc[(projections['name'].isin(fpl_df.loc[my_starting11, 'name'].values)) & 
                (projections['gameweek']==gameweek)].sort_values(by='position'))
print()
print('Substitutes:')
display(projections.loc[(projections['name'].isin(fpl_df.loc[my_subs, 'name'].values)) & 
                (projections['gameweek']==gameweek)].sort_values(by='position'))
```
