```python
aux1 = df_new_season[['whole_name', 'now_cost']].copy()
aux2 = df_previous_season[['whole_name', 'adjusted points per game', 'points_per_game', 'element_type', 
                           'games played']].copy()
display(aux1)
display(aux2)
```

```python
print('previous season:')
display(aux2[aux2['whole_name'].str.contains('Cance')])
print('new season:')
display(aux1[aux1['whole_name'].str.contains('Cance')])
```

```python
# fix some player names that are different between previous and this season, e.g. Benjamin Chilwell -> Ben Chilwell
aux2.loc[135, 'whole_name'] = 'Ben Chilwell'
aux1.loc[306, 'whole_name'] = 'Joao Cancelo'
aux2.loc[256, 'whole_name'] = 'Joao Cancelo'
```

```python
df = aux1.merge(aux2, on='whole_name', how='inner')
df
```

```python
df['price'] = df['now_cost'].astype(int) / 10.0
df
```

```python
df = df.drop(columns=['now_cost'])
df
```

```python
df['value_appg'] = df['adjusted points per game'] / df['price']
df['value_ppg'] = df['points_per_game'] / df['price']
df
```

```python
df['geom_value_appg'] = np.sqrt(df['adjusted points per game'] * df['value_appg'])
df['geom_value_ppg'] = np.sqrt(df['adjusted points per game'] * df['value_ppg'])
df
```

```python
top_n = 10

print('GOALKEEPERS')
print()
print('Sorted by the geometric mean of adjusted points per game and value')
display(df[(df['games played']>=10) & (df['element_type']==1)].sort_values(by='geom_value_appg', 
                                                                           ascending=False).head(top_n))

print()
print('Sorted by the geometric mean of points per game and value')
display(df[(df['games played']>=10) & (df['element_type']==1)].sort_values(by='geom_value_ppg', 
                                                                           ascending=False).head(top_n))
```
