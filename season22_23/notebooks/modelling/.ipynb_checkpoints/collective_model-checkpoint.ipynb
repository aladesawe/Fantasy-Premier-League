```python
test_data = all_data.loc[all_data['season']=='2021-22', keep].copy()
test_data
```

```python
target = all_data.loc[all_data['season']=='2021-22', 'total_points'].copy()
target
```

```python
xgb_models = []
for i in range(1,11):
    xgb = XGBRegressor()
    xgb.load_model(f'../../models/xgb_regression_full_game{i}.json')
    xgb_models.append(xgb)
```

```python
my_predictions = []
for ix in test_data.index:
    model_predictions = []
    for model_ix in range(10):
        model_predictions.append(xgb_models[model_ix].predict(test_data.loc[ix:ix])[0])
    my_predictions.append(np.mean(model_predictions))
my_predictions
```

```python
target - my_predictions
```

```python
rmse_test = mean_squared_error(target, my_predictions, squared=False)
mae_test = mean_absolute_error(target, my_predictions)
r2_test = r2_score(target, my_predictions)
print(f'Test RMSE: {rmse_test}')
print(f'Test MAE: {mae_test}')
print(f'Test R2: {r2_test}')
```
