```python
gameweek = 7
```

```python
import pandas as pd
import numpy as np
import json
```

```python
projections = pd.read_csv(f'../../point_projections/collective_xP_projections_week{gameweek}.csv', index_col=0)
projections
```

```python
# fetch player_dict 
with open('../../data/player_name_dict.txt', 'r') as f:
    player_name_dict_as_text = f.read()
    player_name_dict = json.loads(player_name_dict_as_text)
```

```python
latest_fpl = pd.read_csv(f'../../data/fpl_data/fpl_week_{gameweek}.csv', index_col=0)
latest_fpl['name'] = latest_fpl['name'].map(player_name_dict).fillna(latest_fpl['name'])
# drop Liverpool's Ben Davies
latest_fpl = latest_fpl.drop(latest_fpl[(latest_fpl['name']=='Ben Davies') & (latest_fpl['team']=='Liverpool')].index)
latest_fpl
```

```python
player_names = projections.name.unique()
```

```python
for my_name in player_names:
    my_id = latest_fpl.loc[latest_fpl.name==my_name, 'id'].values[0]
    projections.loc[projections.name==my_name, 'id'] = my_id
    projections['id'] = projections['id'].astype(int)
projections
```

```python
my_columns = [f'{my_week}_Pts' for my_week in range(gameweek+1,gameweek+11)]
my_columns.append('ID')
data = []
for my_name in player_names:
    aux = projections[projections.name==my_name]
    new_data = []
    for my_week in range(gameweek+1,gameweek+11):
        if aux[aux['gameweek'] == my_week].shape[0]==1:
            new_data.append(aux.loc[aux['gameweek'] == my_week, 'xP'].values[0])
        else:
            new_data.append(0)
    my_id = aux['id'].unique()[0]
    new_data.append(my_id)
    data.append(new_data)

df = pd.DataFrame(data, columns=my_columns)
df
```
